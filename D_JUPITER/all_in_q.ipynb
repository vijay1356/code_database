```python
#lab 2 (ii)|  Cardioids: r = a + acos() and r = a  acos()
import numpy as np
import matplotlib.pyplot as plt
import math

plt.axes(projection='polar')

a = 3
rad = np.arange(0, (2 * np.pi), 0.01)

for i in rad:
    r = a + (a * np.cos(i))
    plt.polar(i, r, 'g.')

    r1 = a - (a * np.cos(i))
    plt.polar(i, r1, 'r.')

plt.show()
```

```python
#lab 5 (i)|  Prove that mixed partial derivatives , uxy = uyx for u = exp(x)(xcos(y)  ysin(y)).
from sympy import *

x, y = symbols('x y')

u = exp(x)*(x*cos(y)-y*sin(y))

dux = diff(u, x)
duy = diff(u, y)

duxy = diff(dux, y)
duyx = diff(duy, x)

if duxy == duyx:
    print('Mixed partial derivatives are equal ')
else:
    print('Mixed partial derivatives are not equal ')
```

```python
#Cycloid: x = a(  sin); y = a(1  sin) | lab 2 (iii)
import numpy as np
import matplotlib.pyplot as plt

def cycloid(r):
    x, y = [], []
    for theta in np.linspace(-2 * np.pi, 2 * np.pi, 100):
        x.append(r * (theta - np.sin(theta)))
        y.append(r * (1 - np.cos(theta)))
    plt.plot(x, y)
    plt.show()

cycloid(2)
```

```python
#lab 3 (i) |  Find the angle between the curves r = 4(1 + cost) and r = 5(1  cost)
from sympy import *
r , t = symbols ('r,t')
r1=4*( 1+cos ( t ) ) ;
r2=5*( 1-cos ( t ) ) ;
dr1 = diff ( r1 , t )
dr2 = diff ( r2 , t )
t1=r1/dr1
t2=r2/dr2
q= solve ( r1-r2 , t )
w1=t1 . subs ({t: float ( q[1])})
w2=t2 . subs ({t: float ( q[1])})
y1= atan ( w1 )
y2= atan ( w2 )
w=abs( y1-y2 )
print ('Angle between curves in radians is %0.3f '%( w ) )
```
