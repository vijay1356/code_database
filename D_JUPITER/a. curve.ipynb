```python
# 1. Using modern mathematical tool write a program/code to plot the sine and cosine  curve.
# Import numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values from -10 to 10 with a step of 0.001
x = np.arange(-10, 10, 0.001)

# Create an array of y1 values as the sine of x
y1 = np.sin(x)

# Create an array of y2 values as the cosine of x
y2 = np.cos(x)

# Plot x, y1 and x, y2 on the same figure
plt.plot(x, y1, x, y2)

# Add a title to the figure
plt.title("Sine curve and cosine curve")

# Add labels to the x and y axes
plt.xlabel("Values of x")
plt.ylabel("Values of sin(x) and cos(x)")

# Add a grid to the figure
plt.grid()

# Show the figure
plt.show()
```

```python
#Using modern mathematical tool write a program/code to test the consistency of the equations, x+2y-z=1, 2x+y+4z=2, 3x+3y+4z=1
import numpy as np 
A=np.matrix([[1,2,-1],[2,1,4],[3,3,4]])
B=np.matrix([[1],[2],[1]])
AB=np.concatenate((A,B),axis=1)
rA=np.linalg.matrix_rank(A)
rAB=np.linalg.matrix_rank(AB)
n=A.shape[1]
if(rA==rAB):
     if(rA==n):
          print("the system has unique solution")
          print(np.linalg.solve(A,B))
     else:
          print("the system has infinite solution")
else:
     print("The system of equations is inconsistent ")
```

```python
# 4) Using modern mathematical tool write a program/code to plot the curve 𝑟 = 2|𝑐𝑜𝑠2𝜃|.
from pylab import *
# the number 1000 represents the number of points at which the function will be evaluated. It specifies that the array theta will contain 1000 evenly spaced values between 0 and 2π. This determines the resolution and smoothness of the polar plot.
theta = linspace(0, 2*pi, 1000)
r = 2*abs(cos(2 * theta))
polar(theta, r, 'g')
show()
```

```python
# 6 Using modern mathematical tool write a program/code to find the largest eigen value of 𝐴  
# [
# 1 1 3
# 1 5 1
# 3 1 1
# ] by power method 

import numpy as np

def normalize(x):
    fac = abs(x).max()
    x_n = x / x.max()
    return fac, x_n


x = np.array([1, 1, 1])
a = np.array([[1, 1, 3],
              [1, 5, 1],
              [3, 1, 1]])

for i in range(10):
    x = np.dot(a, x)
    lambda_1, x = normalize(x)
    print('Eigenvalue:', lambda_1)
    print('Eigenvector:', x)
```
