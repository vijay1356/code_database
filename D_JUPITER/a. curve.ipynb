```python
# 1. Using modern mathematical tool write a program/code to plot the sine and cosine  curve.
# Import numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values from -10 to 10 with a step of 0.001
x = np.arange(-10, 10, 0.001)

# Create an array of y1 values as the sine of x
y1 = np.sin(x)

# Create an array of y2 values as the cosine of x
y2 = np.cos(x)

# Plot x, y1 and x, y2 on the same figure
plt.plot(x, y1, x, y2)

# Add a title to the figure
plt.title("Sine curve and cosine curve")

# Add labels to the x and y axes
plt.xlabel("Values of x")
plt.ylabel("Values of sin(x) and cos(x)")

# Add a grid to the figure
plt.grid()

# Show the figure
plt.show()
```
